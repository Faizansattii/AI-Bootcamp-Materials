### ğŸ¯ Understanding Image Tags

Tags specify versions and variants:

- **`python:3.11`** â†’ Full Python 3.11 with all packages (~900MB)
- **`python:3.11-slim`** â†’ Minimal Python 3.11 (~150MB) âœ… Recommended
- **`python:3.11-alpine`** â†’ Smallest (~50MB) but compatibility issues sometimes
- **`python:latest`** â†’ Always pulls newest version (use specific tags in production!)

**Pro Tip:** Always use specific tags (`3.11-slim`) instead of `latest` for reproducibility!

## ğŸ“ Part 6: Running Python Scripts in Containers

Let's run an actual Python script inside a container!

````python
# Create a simple Python script
import os

os.makedirs('/tmp/docker_test', exist_ok=True)

script_content = '''#!/usr/bin/env python3
import sys
import platform

print("ğŸ Python Script Running in Docker!")
print(f"Python Version: {sys.version}")
print(f"Platform: {platform.platform()}")
print(f"Architecture: {platform.machine()}")
print("âœ… Script completed successfully!")
'''

with open('/tmp/docker_test/test_script.py', 'w') as f:
    f.write(script_content)

print("âœ… Created test_script.py")
print(f"ğŸ“ Location: /tmp/docker_test/test_script.py")
````

### ğŸ—‚ï¸ Volume Mounting

**To run the script in a container, we need to mount our local directory:**

**Run this in your terminal:**

```bash
docker run -v /tmp/docker_test:/app python:3.11-slim python /app/test_script.py
```

**Command breakdown:**
- `-v /tmp/docker_test:/app`: Mount local `/tmp/docker_test` to container `/app`
- `python:3.11-slim`: Image to use
- `python /app/test_script.py`: Command to run

**ğŸ’¡ Key Concept:** Volumes let containers access host files!

## ğŸ—ï¸ Part 7: Creating Your First Dockerfile

Dockerfiles are recipes for building custom images. Let's create one!

````python
# Create a simple FastAPI application
app_code = '''from fastapi import FastAPI
import platform

app = FastAPI()

@app.get("/")
def read_root():
    return {
        "message": "Hello from Docker!",
        "python_version": platform.python_version(),
        "platform": platform.platform()
    }

@app.get("/health")
def health_check():
    return {"status": "healthy"}
'''

with open('/tmp/docker_test/app.py', 'w') as f:
    f.write(app_code)

print("âœ… Created app.py")
````

````python
# Create requirements.txt
requirements = '''fastapi==0.104.1
uvicorn==0.24.0
'''

with open('/tmp/docker_test/requirements.txt', 'w') as f:
    f.write(requirements)

print("âœ… Created requirements.txt")
````

````python
# Create Dockerfile
dockerfile = '''# Use official Python image
FROM python:3.11-slim

# Set working directory
WORKDIR /app

# Copy requirements first (for better caching)
COPY requirements.txt .

# Install dependencies
RUN pip install --no-cache-dir -r requirements.txt

# Copy application code
COPY app.py .

# Expose port
EXPOSE 8000

# Run the application
CMD ["uvicorn", "app:app", "--host", "0.0.0.0", "--port", "8000"]
'''

with open('/tmp/docker_test/Dockerfile', 'w') as f:
    f.write(dockerfile)

print("âœ… Created Dockerfile")
print("\nğŸ“‹ Dockerfile contents:")
print(dockerfile)
````

### ğŸ¯ Dockerfile Instructions Explained

- **`FROM python:3.11-slim`**: Start from Python base image
- **`WORKDIR /app`**: Set working directory inside container
- **`COPY requirements.txt .`**: Copy file from host to container
- **`RUN pip install ...`**: Execute command during build
- **`COPY app.py .`**: Copy application code
- **`EXPOSE 8000`**: Document which port the app uses
- **`CMD [...]`**: Default command when container starts

**ğŸ”‘ Key Principle:** Order matters for caching! Put rarely-changing stuff (dependencies) before frequently-changing stuff (code).

### ğŸ—ï¸ Building the Image

**Run these commands in your terminal:**

```bash
cd /tmp/docker_test
docker build -t my-fastapi-app:v1 .
```

**What happens:**
1. Docker reads Dockerfile
2. Executes each instruction
3. Creates layers
4. Tags final image as `my-fastapi-app:v1`

**Then run the container:**

```bash
docker run -d -p 8000:8000 --name fastapi-container my-fastapi-app:v1
```

**Test it:**
```bash
curl http://localhost:8000
curl http://localhost:8000/health
```

## ğŸ§¹ Part 8: Cleanup & Best Practices

Let's clean up our containers and images properly.

````python
# Stop all running containers
print("ğŸ›‘ Stopping all containers...")
!docker stop $(docker ps -q) 2>/dev/null || echo "No containers running"

print("\nâœ… All containers stopped")
````

````python
# Remove all stopped containers
print("ğŸ—‘ï¸  Removing stopped containers...")
!docker container prune -f

print("\nâœ… Stopped containers removed")
````

````python
# Remove unused images
print("ğŸ—‘ï¸  Removing unused images...")
!docker image prune -f

print("\nâœ… Unused images removed")
````

````python
# Check disk space used by Docker
print("ğŸ’¾ Docker disk usage:\n")
!docker system df

print("\nğŸ’¡ Shows space used by images, containers, and volumes")
````

### ğŸ“‹ Docker Best Practices Checklist

âœ… **Use specific image tags** (not `latest`)  
âœ… **Use slim/alpine variants** when possible  
âœ… **Put dependencies before code** in Dockerfile  
âœ… **Use `.dockerignore`** to exclude unnecessary files  
âœ… **Clean up stopped containers** regularly  
âœ… **Name your containers** for easy management  
âœ… **Use volumes** for persistent data  
âœ… **Always specify port mappings** clearly  
âœ… **Check logs** when debugging  
âœ… **One process per container** (microservices principle)
