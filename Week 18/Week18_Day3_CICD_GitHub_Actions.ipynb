## ğŸ¯ Part 1: What is CI/CD?

### ğŸ“š Definitions

#### **Continuous Integration (CI)**
Automatically **build** and **test** your code every time someone pushes changes.

**Benefits:**
- âœ… Catch bugs early (before they reach production)
- âœ… Ensure code quality automatically
- âœ… Faster feedback for developers
- âœ… Consistent build process

#### **Continuous Deployment (CD)**
Automatically **deploy** your code to production after passing all tests.

**Benefits:**
- âœ… Deploy 10x faster (minutes instead of hours)
- âœ… No manual steps (less human error)
- âœ… Reliable, repeatable process
- âœ… Deploy multiple times per day

## ğŸ¤– Part 2: Introduction to GitHub Actions

### ğŸ“š What is GitHub Actions?

**GitHub Actions** is GitHub's built-in CI/CD platform.

**Features:**
- âœ… Free for public repositories
- âœ… 2,000 minutes/month free for private repos
- âœ… No extra tools to install
- âœ… Workflows defined in YAML files
- âœ… Runs on GitHub's servers (no need for your own)

### ğŸ“ Workflow Structure

Workflows are stored in `.github/workflows/` directory as YAML files.

```yaml
name: CI/CD Pipeline          # Workflow name

on:                           # When to run
  push:
    branches: [main]

jobs:                         # Tasks to run
  build:                      # Job name
    runs-on: ubuntu-latest    # Environment
    steps:                    # Commands to execute
      - uses: actions/checkout@v3
      - name: Build
        run: docker build -t app .
```

### ğŸ§© Key Components

| Component | Purpose | Example |
|-----------|---------|--------|
| **name** | Workflow name | `CI/CD Pipeline` |
| **on** | Trigger events | `push`, `pull_request`, `schedule` |
| **jobs** | Tasks to run | `build`, `test`, `deploy` |
| **runs-on** | OS environment | `ubuntu-latest`, `windows-latest` |
| **steps** | Individual commands | `checkout`, `build`, `test` |
| **uses** | Reusable actions | `actions/checkout@v3` |
| **run** | Shell commands | `docker build`, `npm test` |

### ğŸ’¡ Workflow Triggers

```yaml
# Trigger on push to main
on:
  push:
    branches: [main]

# Trigger on pull request
on:
  pull_request:
    branches: [main]

# Trigger on schedule (daily at 2 AM)
on:
  schedule:
    - cron: '0 2 * * *'

# Trigger manually
on:
  workflow_dispatch:

# Multiple triggers
on:
  push:
    branches: [main]
  pull_request:
    branches: [main]
```

### ğŸ” Why Secrets?

**âŒ NEVER do this:**
```yaml
env:
  DOCKER_USER: john_doe
  DOCKER_PASS: mypassword123  # Visible to everyone!
```

**âœ… DO this:**
```yaml
env:
  DOCKER_USER: ${{ secrets.DOCKER_USERNAME }}
  DOCKER_PASS: ${{ secrets.DOCKER_PASSWORD }}  # Encrypted!
```

Secrets are **encrypted** and **never** shown in logs.

### ğŸ› ï¸ Exercise 4: Create Workflow File

**Step 1: Create workflow directory**

```bash
# ğŸ–¥ï¸ Run in Terminal:
mkdir -p .github/workflows
```

**Step 2: Create `.github/workflows/docker-build.yml`**

```yaml
name: Build and Push Docker Image

on:
  push:
    branches: [main]
  pull_request:
    branches: [main]

env:
  IMAGE_NAME: cicd-demo

jobs:
  build-and-push:
    runs-on: ubuntu-latest
    
    steps:
      # Step 1: Checkout code
      - name: Checkout code
        uses: actions/checkout@v3
      
      # Step 2: Login to Docker Hub
      - name: Login to Docker Hub
        uses: docker/login-action@v2
        with:
          username: ${{ secrets.DOCKER_USERNAME }}
          password: ${{ secrets.DOCKER_PASSWORD }}
      
      # Step 3: Build and push Docker image
      - name: Build and push
        uses: docker/build-push-action@v4
        with:
          context: .
          push: true
          tags: |
            ${{ secrets.DOCKER_USERNAME }}/${{ env.IMAGE_NAME }}:latest
            ${{ secrets.DOCKER_USERNAME }}/${{ env.IMAGE_NAME }}:${{ github.sha }}
```

**Step 3: Commit and push**

```bash
# ğŸ–¥ï¸ Run in Terminal:
git add .github/workflows/docker-build.yml
git commit -m "Add Docker build workflow"
git push
```

**Step 4: Watch it run!**

### ğŸ› ï¸ Exercise 5: Add Tests

**Step 1: Create `test_app.py`**

```python
from fastapi.testclient import TestClient
from app import app

client = TestClient(app)

def test_home():
    response = client.get("/")
    assert response.status_code == 200
    assert "message" in response.json()
    assert response.json()["message"] == "CI/CD Demo App"

def test_health():
    response = client.get("/health")
    assert response.status_code == 200
    assert response.json()["status"] == "healthy"

def test_version():
    response = client.get("/")
    assert "version" in response.json()
```

âœ… All tests pass!

### ğŸ› ï¸ Exercise 6: Add Testing to Workflow

**Update `.github/workflows/docker-build.yml`:**

```yaml
name: Build, Test, and Push

on:
  push:
    branches: [main]
  pull_request:
    branches: [main]

env:
  IMAGE_NAME: cicd-demo

jobs:
  # ========================================
  # JOB 1: Test
  # ========================================
  test:
    runs-on: ubuntu-latest
    
    steps:
      - name: Checkout code
        uses: actions/checkout@v3
      
      - name: Set up Python
        uses: actions/setup-python@v4
        with:
          python-version: '3.11'
      
      - name: Install dependencies
        run: |
          pip install --upgrade pip
          pip install -r requirements.txt
      
      - name: Run tests
        run: pytest test_app.py -v
      
      - name: Lint code
        run: |
          pip install flake8
          flake8 app.py --max-line-length=100
  
  # ========================================
  # JOB 2: Build and Push (only if tests pass!)
  # ========================================
  build-and-push:
    needs: test  # â† Only runs if test job succeeds!
    runs-on: ubuntu-latest
    
    steps:
      - name: Checkout code
        uses: actions/checkout@v3
      
      - name: Login to Docker Hub
        uses: docker/login-action@v2
        with:
          username: ${{ secrets.DOCKER_USERNAME }}
          password: ${{ secrets.DOCKER_PASSWORD }}
      
      - name: Build and push
        uses: docker/build-push-action@v4
        with:
          context: .
          push: true
          tags: |
            ${{ secrets.DOCKER_USERNAME }}/${{ env.IMAGE_NAME }}:latest
            ${{ secrets.DOCKER_USERNAME }}/${{ env.IMAGE_NAME }}:${{ github.sha }}
```

**Key Addition:**
```yaml
needs: test  # This line ensures build only runs if tests pass!
```

**Commit and push:**

```bash
# ğŸ–¥ï¸ Run in Terminal:
git add .
git commit -m "Add automated testing"
git push
```

ğŸŒ **Watch on GitHub Actions tab:**

## â­ Part 5: CI/CD Best Practices

### ğŸ¯ 10 Best Practices

#### 1ï¸âƒ£ **Test Before Deploy**
```yaml
jobs:
  test:
    # ... tests ...
  
  deploy:
    needs: test  # â† CRITICAL!
```
Never deploy untested code!

#### 2ï¸âƒ£ **Use Secrets for Credentials**
```yaml
# âŒ NEVER
username: john_doe

# âœ… ALWAYS
username: ${{ secrets.DOCKER_USERNAME }}
```

#### 3ï¸âƒ£ **Tag Images Properly**
```yaml
tags: |
  myapp:latest
  myapp:${{ github.sha }}     # Git commit SHA
  myapp:v${{ github.run_number }}  # Version number
```
Multiple tags help with rollbacks!

#### 4ï¸âƒ£ **Cache Dependencies**
```yaml
- name: Cache pip packages
  uses: actions/cache@v3
  with:
    path: ~/.cache/pip
    key: ${{ runner.os }}-pip-${{ hashFiles('requirements.txt') }}
```
Speeds up builds significantly!

#### 5ï¸âƒ£ **Fail Fast**
```yaml
jobs:
  test:
    steps:
      - run: pytest
        continue-on-error: false  # Stop immediately on failure
```

#### 6ï¸âƒ£ **Separate Workflows for Dev/Prod**
```
.github/workflows/
  dev.yml        # Runs on dev branch
  staging.yml    # Runs on staging branch
  production.yml # Runs on main branch
```

#### 7ï¸âƒ£ **Monitor and Get Notified**
```yaml
- name: Notify on failure
  if: failure()
  uses: slackapi/slack-github-action@v1
  with:
    payload: '{"text": "Build failed!"}'
```

#### 8ï¸âƒ£ **Use Matrix Builds**
```yaml
strategy:
  matrix:
    python-version: ['3.10', '3.11', '3.12']
    os: [ubuntu-latest, windows-latest]
```
Test on multiple versions/platforms!

#### 9ï¸âƒ£ **Document Everything**
```markdown
# README.md
## CI/CD Pipeline
- Triggers: Push to main
- Tests: pytest
- Deploys to: Docker Hub
```

#### ğŸ”Ÿ **Review Logs When Things Fail**
Always check GitHub Actions logs:
- Repo â†’ Actions tab
- Click failed workflow
- Expand failed step
- Read error message

## ğŸ”§ Troubleshooting

### Common Issues and Solutions:

#### Issue 1: "Workflow not triggering"
**Solution:**
- Check file is in `.github/workflows/` (note the dot!)
- Check YAML syntax (indentation matters!)
- Check trigger matches your branch name
- Go to Actions tab â†’ Click "Enable Actions"

#### Issue 2: "Login to Docker Hub failed"
**Solution:**
- Check secrets are set: Settings â†’ Secrets and variables â†’ Actions
- Secret names match exactly: `DOCKER_USERNAME`, `DOCKER_PASSWORD`
- Try using Docker Hub access token instead of password
- Verify credentials work locally: `docker login`

#### Issue 3: "Tests passed but build didn't run"
**Solution:**
- Check `needs: test` is in build job
- Check indentation is correct
- Look for YAML syntax errors in Actions tab

#### Issue 4: "Build failed: unable to prepare context"
**Solution:**
- Check Dockerfile is in repository root
- Check `context: .` in workflow
- Make sure you pushed Dockerfile to GitHub

#### Issue 5: "Push to Docker Hub permission denied"
**Solution:**
- Check Docker Hub username is lowercase
- Verify repository name doesn't have uppercase
- Format: `username/repository:tag` (all lowercase)
