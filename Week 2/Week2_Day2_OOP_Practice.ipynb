## Part 6: Inheritance (Basics)

### What is Inheritance?
One class can **inherit** (get) properties and methods from another class.

**Example:**
- Parent class: Animal
- Child class: Dog (gets everything from Animal + adds its own)

### Example: Animal and Dog

````python
# Example: Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    
    def eat(self):
        return f"{self.name} is eating"

# Dog inherits from Animal
class Dog(Animal):
    def bark(self):
        return f"{self.name} says Woof!"

# Create dog
dog1 = Dog("Buddy")

# Dog has both Animal methods AND its own methods
print(dog1.eat())   # From Animal class
print(dog1.bark())  # From Dog class
````

## Part 7: Encapsulation (Basics)

### What is Encapsulation?
Encapsulation means **hiding** some data and only allowing access through methods.

We use:
- `_` (one underscore) = "Don't touch this directly, please"
- `__` (two underscores) = "Really private, use methods only"

### Example: Bank account with private balance

````python
# Example: Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private (note: __)
    
    def deposit(self, amount):
        self.__balance = self.__balance + amount
    
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance = self.__balance - amount
        else:
            print("Not enough money!")
    
    def get_balance(self):
        return self.__balance

# Create account
account = BankAccount(1000)

print(f"Balance: ${account.get_balance()}")

account.deposit(500)
print(f"After deposit: ${account.get_balance()}")

account.withdraw(200)
print(f"After withdraw: ${account.get_balance()}")
````
