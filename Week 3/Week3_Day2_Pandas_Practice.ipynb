### Exercise 4: Handling Missing Values
**Learn:** Detect and handle missing data (NaN values) in DataFrames.

#### ğŸ“– Example:

```python
# Example: Handling missing values

# Create DataFrame with missing values
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, np.nan, 35, 28, np.nan],
    'Salary': [50000, 60000, np.nan, 55000, 65000]
})
print("DataFrame with missing values:")
print(df)

# Check for missing values
print("\nMissing values (True/False):")
print(df.isnull())

# Count missing values per column
print("\nMissing value count:")
print(df.isnull().sum())

# Drop rows with any missing values
print("\nAfter dropping rows with NaN:")
print(df.dropna())

# Fill missing values with a specific value
print("\nFill Age NaN with 30, Salary NaN with 0:")
df_filled = df.fillna({'Age': 30, 'Salary': 0})
print(df_filled)

# Fill with mean
print("\nFill missing Age with mean:")
df['Age'].fillna(df['Age'].mean(), inplace=False)
print(df)

# Expected Output:
# DataFrame with missing values:
#       Name   Age   Salary
# 0    Alice  25.0  50000.0
# 1      Bob   NaN  60000.0
# 2  Charlie  35.0      NaN
# 3    David  28.0  55000.0
# 4      Eve   NaN  65000.0
```

### Exercise 5: Data Transformation with apply, map, and lambda
**Learn:** Transform data using functions.

#### ğŸ“– Example:

```python
# Example: Data transformation

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 75000]
})
print("Original DataFrame:")
print(df)

# Using lambda with apply (on a column)
df['Salary_Increased'] = df['Salary'].apply(lambda x: x * 1.10)
print("\nAfter 10% salary increase:")
print(df)

# Using a custom function
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif age < 40:
        return 'Mid'
    else:
        return 'Senior'

df['Age_Category'] = df['Age'].apply(categorize_age)
print("\nWith age categories:")
print(df)

# Using map to replace values
name_map = {'Alice': 'A', 'Bob': 'B', 'Charlie': 'C'}
df['Name_Code'] = df['Name'].map(name_map)
print("\nWith name codes:")
print(df)

# Apply to entire DataFrame
print("\nDouble all numeric columns:")
numeric_doubled = df[['Age', 'Salary']].apply(lambda x: x * 2)
print(numeric_doubled)

# Expected Output:
# After 10% salary increase:
#       Name  Age  Salary  Salary_Increased
# 0    Alice   25   50000           55000.0
# 1      Bob   30   60000           66000.0
# 2  Charlie   35   75000           82500.0
```

### Exercise 7: Merging and Joining DataFrames
**Learn:** Combine multiple DataFrames based on common columns.

#### ğŸ“– Example:

```python
# Example: Merging DataFrames

# Create two DataFrames
employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'DepartmentID': [101, 102, 101, 103]
})

departments = pd.DataFrame({
    'DepartmentID': [101, 102, 103],
    'Department': ['IT', 'HR', 'Finance']
})

print("Employees:")
print(employees)
print("\nDepartments:")
print(departments)

# Inner join (only matching rows)
merged_inner = pd.merge(employees, departments, on='DepartmentID', how='inner')
print("\nInner Join:")
print(merged_inner)

# Left join (all rows from left, matching from right)
merged_left = pd.merge(employees, departments, on='DepartmentID', how='left')
print("\nLeft Join:")
print(merged_left)

# Concatenate DataFrames vertically
more_employees = pd.DataFrame({
    'EmployeeID': [5, 6],
    'Name': ['Eve', 'Frank'],
    'DepartmentID': [102, 101]
})

all_employees = pd.concat([employees, more_employees], ignore_index=True)
print("\nConcatenated (vertical):")
print(all_employees)

# Expected Output:
# Inner Join:
#    EmployeeID     Name  DepartmentID Department
# 0           1    Alice           101         IT
# 1           2      Bob           102         HR
# 2           3  Charlie           101         IT
# 3           4    David           103    Finance
```

### Exercise 8: GroupBy and Aggregation
**Learn:** Group data and perform aggregate operations.

#### ğŸ“– Example:

```python
# Example: GroupBy operations

df = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Salary': [70000, 60000, 75000, 80000, 65000, 72000, 78000],
    'Experience': [3, 5, 4, 7, 6, 3, 5]
})

print("DataFrame:")
print(df)

# Group by Department and calculate mean salary
print("\nAverage Salary by Department:")
print(df.groupby('Department')['Salary'].mean())

# Multiple aggregations
print("\nMultiple Aggregations:")
print(df.groupby('Department').agg({
    'Salary': ['mean', 'min', 'max'],
    'Experience': 'mean'
}))

# Count employees per department
print("\nEmployee Count by Department:")
print(df.groupby('Department').size())

# Group by and create new DataFrame
dept_summary = df.groupby('Department').agg({
    'Salary': 'mean',
    'Experience': 'mean',
    'Employee': 'count'
}).reset_index()
dept_summary.columns = ['Department', 'Avg_Salary', 'Avg_Experience', 'Employee_Count']
print("\nDepartment Summary:")
print(dept_summary)

# Expected Output:
# Average Salary by Department:
# Department
# Finance    79000.0
# HR         62500.0
# IT         72333.333333
```
