**Exploratory Data Analysis (EDA)** is the process of:
- Understanding your data structure and contents
- Identifying patterns, trends, and anomalies
- Discovering relationships between variables
- Detecting outliers and missing values
- Preparing data for machine learning

**Why EDA?**
- ğŸ“Š **Understand Data**: Know what you're working with
- ğŸ” **Find Issues**: Detect problems early
- ğŸ’¡ **Generate Insights**: Discover hidden patterns

```python
# Example: Summary statistics

# Create sample data
scores = pd.Series([85, 92, 78, 95, 88, 76, 90, 85, 82, 89])

print("Scores:", scores.values)
print("\nSummary Statistics:")
print(f"Mean: {scores.mean():.2f}")
print(f"Median: {scores.median():.2f}")
print(f"Mode: {scores.mode().values[0]}")
print(f"Standard Deviation: {scores.std():.2f}")
print(f"Variance: {scores.var():.2f}")
print(f"Min: {scores.min()}")
print(f"Max: {scores.max()}")
print(f"Range: {scores.max() - scores.min()}")
print(f"25th Percentile (Q1): {scores.quantile(0.25):.2f}")
print(f"75th Percentile (Q3): {scores.quantile(0.75):.2f}")
print(f"IQR: {scores.quantile(0.75) - scores.quantile(0.25):.2f}")

# Expected Output:
# Mean: 86.00
# Median: 86.50
# Standard Deviation: 5.89
```

### Exercise 5: Box Plots for Outlier Detection
**Learn:** Use box plots to identify outliers and understand data spread.

#### ğŸ“– Example:

```python
# Example: Box plot

# Create sample data with outliers
np.random.seed(42)
data = {
    'Group_A': np.concatenate([np.random.normal(50, 10, 95), [5, 10, 90, 95, 100]]),
    'Group_B': np.concatenate([np.random.normal(60, 8, 95), [10, 15, 95, 100, 105]]),
    'Group_C': np.concatenate([np.random.normal(55, 12, 95), [8, 12, 92, 98, 102]])
}
df = pd.DataFrame(data)

# Create box plots
plt.figure(figsize=(12, 5))

# Matplotlib box plot
plt.subplot(1, 2, 1)
plt.boxplot([df['Group_A'], df['Group_B'], df['Group_C']], 
            labels=['Group A', 'Group B', 'Group C'])
plt.ylabel('Values')
plt.title('Box Plot - Matplotlib')
plt.grid(True, alpha=0.3)

# Seaborn box plot
plt.subplot(1, 2, 2)
sns.boxplot(data=df, palette='Set2')
plt.ylabel('Values')
plt.title('Box Plot - Seaborn')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Box plot components:")
print("- Box: Interquartile range (IQR) - middle 50% of data")
print("- Line in box: Median")
print("- Whiskers: 1.5 * IQR from quartiles")
print("- Points outside whiskers: Potential outliers")
```

### Exercise 6: Scatter Plots and Correlation
**Learn:** Visualize relationships between two variables.

#### ğŸ“– Example:

```python
# Example: Scatter plot and correlation

# Create correlated data
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5  # y correlated with x

# Calculate correlation
correlation = np.corrcoef(x, y)[0, 1]

# Create scatter plot
plt.figure(figsize=(12, 5))

# Matplotlib scatter
plt.subplot(1, 2, 1)
plt.scatter(x, y, alpha=0.6, color='blue')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title(f'Scatter Plot (Correlation: {correlation:.3f})')
plt.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", alpha=0.8, linewidth=2, label='Trend line')
plt.legend()

# Seaborn scatter with regression line
plt.subplot(1, 2, 2)
sns.regplot(x=x, y=y, scatter_kws={'alpha': 0.6})
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title(f'Scatter Plot with Regression (Seaborn)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nCorrelation coefficient: {correlation:.3f}")
print("Interpretation:")
print("  > 0.7: Strong positive correlation")
print("  0.3 to 0.7: Moderate positive correlation")
print("  -0.3 to 0.3: Weak or no correlation")
print("  -0.7 to -0.3: Moderate negative correlation")
print("  < -0.7: Strong negative correlation")
```

### Exercise 7: Correlation Heatmaps
**Learn:** Visualize correlations between multiple variables using heatmaps.

#### ğŸ“– Example:

```python
# Example: Correlation heatmap

# Create sample dataset
np.random.seed(42)
n = 100
data = pd.DataFrame({
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(30000, 100000, n),
    'Experience': np.random.randint(0, 30, n),
    'Education': np.random.randint(12, 20, n)
})

# Add some correlation
data['Income'] = data['Experience'] * 2000 + np.random.randint(20000, 50000, n)

# Calculate correlation matrix
corr_matrix = data.corr()

print("Correlation Matrix:")
print(corr_matrix)

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True,  # Show correlation values
            cmap='coolwarm',  # Color scheme
            center=0,  # Center colormap at 0
            square=True,  # Make cells square
            linewidths=1,  # Add lines between cells
            cbar_kws={"shrink": 0.8},  # Colorbar size
            fmt='.2f')  # Format numbers to 2 decimal places

plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nInterpretation:")
print("- Red colors: Positive correlation")
print("- Blue colors: Negative correlation")
print("- White: No correlation")
```
