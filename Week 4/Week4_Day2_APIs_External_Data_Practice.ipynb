**APIs (Application Programming Interfaces)** allow programs to communicate:
- Access data from external services
- Integrate third-party functionality
- Build data pipelines from web sources
- Automate data collection

**Why APIs for AI/ML?**
- ğŸŒ **Access Real Data**: Live data from real services
- ğŸ”„ **Automation**: Automatic data collection
- ğŸ“Š **Rich Sources**: Weather, news, social media, finance
- ğŸš€ **Scalability**: Handle large data volumes

**REST API Basics:**
- **HTTP Methods**: GET (read), POST (create), PUT (update), DELETE
- **Endpoints**: URLs that provide specific data
- **Parameters**: Customize requests (query params, headers)
- **Response**: Usually JSON format
- **Status Codes**: 200 (success), 404 (not found), 429 (rate limit)

### Exercise 4: Error Handling and Status Codes
**Learn:** Properly handle API errors and different response codes.

#### ğŸ“– Example:

````python
# Example: Comprehensive error handling

def fetch_data_safely(url, params=None):
    """
    Safely fetch data from API with error handling
    """
    try:
        # Set timeout to avoid hanging
        response = requests.get(url, params=params, timeout=10)
        
        # Check status code
        if response.status_code == 200:
            print("âœ… Success!")
            return response.json()
        elif response.status_code == 404:
            print("âŒ Error 404: Resource not found")
            return None
        elif response.status_code == 429:
            print("âš ï¸ Error 429: Rate limit exceeded")
            return None
        elif response.status_code >= 500:
            print(f"âŒ Server Error {response.status_code}")
            return None
        else:
            print(f"âš ï¸ Unexpected status code: {response.status_code}")
            return None
            
    except requests.exceptions.Timeout:
        print("âŒ Request timed out")
        return None
    except requests.exceptions.ConnectionError:
        print("âŒ Connection error")
        return None
    except requests.exceptions.RequestException as e:
        print(f"âŒ Request failed: {e}")
        return None
    except json.JSONDecodeError:
        print("âŒ Invalid JSON response")
        return None

# Test with valid URL
print("Test 1: Valid URL")
data = fetch_data_safely('https://jsonplaceholder.typicode.com/posts/1')
if data:
    print(f"Received data with title: {data['title']}")

# Test with invalid resource (404)
print("\nTest 2: Invalid resource")
data = fetch_data_safely('https://jsonplaceholder.typicode.com/posts/99999')

# Test with invalid URL
print("\nTest 3: Invalid URL")
data = fetch_data_safely('https://invalid-domain-12345.com/api')

print("\nâœ… Error handling demonstration complete")
````

### Exercise 5: Rate Limiting and Retries
**Learn:** Implement rate limiting and retry logic for robust API calls.

#### ğŸ“– Example:

````python
# Example: Rate limiting and exponential backoff

def fetch_with_retry(url, max_retries=3, initial_delay=1):
    """
    Fetch data with exponential backoff retry logic
    """
    for attempt in range(max_retries):
        try:
            print(f"Attempt {attempt + 1}/{max_retries}...")
            response = requests.get(url, timeout=5)
            
            if response.status_code == 200:
                print("âœ… Success!")
                return response.json()
            elif response.status_code == 429:
                # Rate limit hit - wait longer
                wait_time = initial_delay * (2 ** attempt)
                print(f"âš ï¸ Rate limited. Waiting {wait_time} seconds...")
                time.sleep(wait_time)
            else:
                print(f"âŒ Error {response.status_code}")
                return None
                
        except requests.exceptions.RequestException as e:
            print(f"âŒ Request failed: {e}")
            if attempt < max_retries - 1:
                wait_time = initial_delay * (2 ** attempt)
                print(f"Retrying in {wait_time} seconds...")
                time.sleep(wait_time)
    
    print("âŒ All retries failed")
    return None

# Test with rate limiting
print("Fetching multiple posts with rate limiting:")
print("="*50)

urls = [
    'https://jsonplaceholder.typicode.com/posts/1',
    'https://jsonplaceholder.typicode.com/posts/2',
    'https://jsonplaceholder.typicode.com/posts/3'
]

results = []
for i, url in enumerate(urls, 1):
    print(f"\nFetching post {i}:")
    data = fetch_with_retry(url)
    if data:
        results.append(data)
    
    # Rate limiting: wait between requests
    if i < len(urls):
        print("â³ Waiting 0.5 seconds before next request...")
        time.sleep(0.5)

print(f"\nâœ… Successfully fetched {len(results)} posts")
````

### Exercise 7: Pagination - Fetching Large Datasets
**Learn:** Handle paginated API responses to get all data.

#### ğŸ“– Example:

````python
# Example: Pagination with offset/limit

def fetch_all_posts(limit_per_page=10):
    """
    Fetch all posts using pagination
    """
    url = 'https://jsonplaceholder.typicode.com/posts'
    all_posts = []
    page = 1
    
    while True:
        # Calculate offset
        start = (page - 1) * limit_per_page
        
        params = {
            '_start': start,
            '_limit': limit_per_page
        }
        
        print(f"Fetching page {page} (items {start} to {start + limit_per_page})...")
        
        response = requests.get(url, params=params)
        
        if response.status_code != 200:
            print(f"âŒ Error: {response.status_code}")
            break
        
        posts = response.json()
        
        if not posts:  # No more data
            print("âœ… Reached end of data")
            break
        
        all_posts.extend(posts)
        print(f"   Retrieved {len(posts)} posts (total: {len(all_posts)})")
        
        # Safety limit
        if page >= 20:  # Prevent infinite loops
            print("âš ï¸ Reached page limit")
            break
        
        page += 1
        time.sleep(0.3)  # Rate limiting
    
    return all_posts

# Fetch all posts
print("Fetching all posts with pagination:")
print("="*50)
all_posts = fetch_all_posts(limit_per_page=20)

print(f"\nâœ… Total posts fetched: {len(all_posts)}")
print(f"First post: {all_posts[0]['title'] if all_posts else 'None'}")
print(f"Last post: {all_posts[-1]['title'] if all_posts else 'None'}")
````

## ğŸ”¹ Section 5: Summary & Key Concepts

### ğŸ¯ What You Learned Today:

**1. HTTP Requests with `requests` Library:**
- `requests.get()`: Fetch data from API
- `response.status_code`: Check request status
- `response.json()`: Parse JSON response
- `response.headers`: Access response headers
- `response.elapsed`: Measure response time

**2. REST API Structure:**
- **Endpoints**: Specific URLs for resources
- **Methods**: GET, POST, PUT, DELETE
- **Parameters**: Query params, path params
- **Headers**: Metadata, authentication
- **Response**: Usually JSON format

**3. Query Parameters:**
- Pass data in URL: `?key=value&key2=value2`
- Use `params` dict: `requests.get(url, params={...})`
- Common params: limit, offset, filter, sort

**4. Headers:**
- **User-Agent**: Identify your application
- **Accept**: Specify response format
- **Authorization**: API keys, tokens
- **Content-Type**: Request body format

**5. HTTP Status Codes:**
- **2xx**: Success (200 OK, 201 Created)
- **4xx**: Client error (400 Bad Request, 404 Not Found, 429 Too Many Requests)
- **5xx**: Server error (500 Internal Server Error, 503 Service Unavailable)

**6. Error Handling:**
- Try-except blocks for network errors
- Check status codes
- Handle timeouts: `timeout` parameter
- Validate JSON responses
- Implement retry logic

**7. Rate Limiting:**
- Respect API rate limits
- Add delays between requests: `time.sleep()`
- Exponential backoff for retries
- Check rate limit headers
- Monitor API usage

**8. Pagination:**
- **Offset-based**: `?offset=0&limit=10`
- **Page-based**: `?page=1&per_page=10`
- **Cursor-based**: `?cursor=xyz&limit=10`
- Loop until no more data
- Aggregate results

**9. Data Processing:**
- Parse JSON to Python objects
- Convert to Pandas DataFrame
- Clean and normalize data
- Handle nested structures
- Save in multiple formats
